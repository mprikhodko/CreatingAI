Using the Amazon musical instrument review data on KaggleLinks to an external site., retrieve the summary column, perform tokenization, stemming, and lemmatization. Construct results of text processing as an jupyter notebook.

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
df = pd.read_csv('Musical_instruments_reviews.csv')
```

```python
df
```

Lemmatization is the process of finding the form of the related word in the dictionary  WordNetLemmatizer() was used. Stemming()Stemming is the process of finding the root of words. Word

```python
# Create a copy of the original DataFrame
df_out = df.copy()

# Tokenize the 'summary' column
df_out['TokenizedSummary'] = df_out['summary'].apply(lambda x: word_tokenize(x.lower()))

# Perform stemming on the 'Tokenized Summary' column
stemmer = PorterStemmer()
df_out['StemmedSummary'] = df_out['TokenizedSummary'].apply(lambda x: [stemmer.stem(word) for word in x])

# Perform lemmatization on the 'Tokenized Summary' column
lemmatizer = WordNetLemmatizer()
df_out['LemmatizedSummary'] = df_out['TokenizedSummary'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
df_out['LemmatizedSummaryAfterStem'] = df_out['TokenizedSummary'].apply(lambda x: [lemmatizer.lemmatize(stemmer.stem(word)) for word in x])

# Print the updated DataFrame
df_out[['summary','TokenizedSummary','StemmedSummary','LemmatizedSummary','LemmatizedSummaryAfterStem']]
```

```python
column_names = ['TokenizedSummary', 'StemmedSummary', 'LemmatizedSummary','LemmatizedSummaryAfterStem']

# Apply describe() function using lambda in-line
statistics = df_out[column_names].apply(lambda x: x.describe())

# Print the descriptive statistics
print(statistics)
```

```python
df_out.to_csv("output.csv")
```
